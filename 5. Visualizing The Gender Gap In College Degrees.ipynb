# Visualizing The Gender Gap In College Degrees

In this project, we will plot few charts to compare the gender gaps in college degrees. This project in not so much about analyzing data, but more about applying skills of creating nice plot learned in Dataquest course.

Degrees we are going to analyze are split in three categories, so it was assumed that it would be nice to represent each category in seperate column:
- STEM
- Library/Arts
- Other

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('data/percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

def generate_subplots(cats_list, column):
    for sp in range(0, len(cats_list)):
        
        ax = fig.add_subplot(6,3,column)
        ax.plot(women_degrees['Year'], women_degrees[cats_list[sp]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[cats_list[sp]], c=cb_orange, label='Men', linewidth=3)
        ax.spines["right"].set_visible(False)    
        ax.spines["top"].set_visible(False)    
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_yticks([0, 25, 50, 75, 100])
        ax.set_xticks([1968, 1982, 1996, 2011])
        ax.set_title(cats_list[sp])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        if sp == 0:
            ax.legend(['Women', 'Men'], loc='upper right')
        if sp == len(cats_list) - 1:
            ax.legend(['Women', 'Men'], loc='upper right')
        column += 3

fig = plt.figure(figsize=(18, 30))
generate_subplots(stem_cats, 1)
generate_subplots(lib_arts_cats, 2)
generate_subplots(other_cats, 3)
# plt.savefig('degree_gaps.png')
plt.show()
```
